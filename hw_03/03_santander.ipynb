```python
import pandas as pd
```

```python
train = pd.read_csv('03_santander_train.csv', index_col='ID')
```

```python
train.head()
```

```python
test = pd.read_csv('03_santander_test.csv', index_col='ID')
```

```python
test.head()
```

```python
y = list(train['TARGET'])
```

```python
train = train.drop('TARGET', axis=1)
```

```python
train.info()
```

```python
train.info(verbose=True, memory_usage=True, null_counts=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(max_depth=5)
dtc.fit(X_train, y_train)
```

```python
from ipywidgets import Image
from io import StringIO
import pydotplus
from sklearn.tree import export_graphviz

dot_data = StringIO()
export_graphviz(dtc, feature_names=train.columns, out_file=dot_data, filled=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(value=graph.create_png())
```

```python
# построим предсказание для max_depth=5

dtc_05 = dtc.predict_proba(X_test)
print(dtc_05)
```

```python
# импортируем метрики

from sklearn.metrics import roc_auc_score, roc_curve

# и вычислим их для max_depth=5

fpr_05, tpr_05, thresholds_05 = roc_curve(y_test, dtc_05[:,1])
```

```python
# те же самые действия для max_depth = 1, 10, 30, 100, 200, 300, 500 
```

```python
dtc = DecisionTreeClassifier(max_depth=1)
dtc.fit(X_train, y_train)
dtc_01 = dtc.predict_proba(X_test)
fpr_01, tpr_01, thresholds_01 = roc_curve(y_test, dtc_01[:,1])
```

```python
dtc = DecisionTreeClassifier(max_depth=10)
dtc.fit(X_train, y_train)
dtc_10 = dtc.predict_proba(X_test)
fpr_10, tpr_10, thresholds_10 = roc_curve(y_test, dtc_10[:,1])
```

```python
dtc = DecisionTreeClassifier(max_depth=30)
dtc.fit(X_train, y_train)
dtc_50 = dtc.predict_proba(X_test)
fpr_50, tpr_50, thresholds_50 = roc_curve(y_test, dtc_50[:,1])
```

```python
dtc = DecisionTreeClassifier(max_depth=100)
dtc.fit(X_train, y_train)
dtc_100 = dtc.predict_proba(X_test)
fpr_100, tpr_100, thresholds_100 = roc_curve(y_test, dtc_100[:,1])
```

```python
dtc = DecisionTreeClassifier(max_depth=200)
dtc.fit(X_train, y_train)
dtc_200 = dtc.predict_proba(X_test)
fpr_200, tpr_200, thresholds_200 = roc_curve(y_test, dtc_200[:,1])
```

```python
dtc = DecisionTreeClassifier(max_depth=300)
dtc.fit(X_train, y_train)
dtc_300 = dtc.predict_proba(X_test)
fpr_300, tpr_300, thresholds_300 = roc_curve(y_test, dtc_300[:,1])
```

```python
dtc = DecisionTreeClassifier(max_depth=500)
dtc.fit(X_train, y_train)
dtc_500 = dtc.predict_proba(X_test)
fpr_500, tpr_500, thresholds_500 = roc_curve(y_test, dtc_500[:,1])
```

```python
# Построим график AUC ROC от красного при глубине один, до синего при глубине 369
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(8, 8))
plt.plot(fpr_01, tpr_01, color='red')
plt.plot(fpr_05, tpr_05, color='#DB0024')
plt.plot(fpr_10, tpr_10, color='#B60049')
plt.plot(fpr_50, tpr_50, color='#92006D')
plt.plot(fpr_100, tpr_100, color='#6D0092')
plt.plot(fpr_200, tpr_200, color='#4900B6')
plt.plot(fpr_300, tpr_300, color='#2400DB')
plt.plot(fpr_500, tpr_500, color='blue')
plt.plot([0, 1], [0, 1], color='y')
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
```

Видим, что площадь под ROC кривой распределена неравномерно. 
Минимальные значения при большей глубине деревьев (от 100),
а максимальные — при значениях в районе 5-10 узлов.
Причем значения достаточно близки друг к другу.
При значении глубины в 30, площадь близка к глубине один.
